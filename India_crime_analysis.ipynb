```python
df_keys.sort_values(by="srcYear",inplace = True)
df_Ndap_report.sort_values(by="Yearcode",inplace = True)
df_source_data.sort_values(by="srcYear",inplace = True)
```

```python
def func_execute(dataframe,state,column):
    filt = dataframe["srcStateName"] == state
    df = dataframe[filt].loc[:,:]
    return df[column]
```

```python
# Registered murder cases in the year of [2016,2017,2018] in State [Delhi,Bihar,Karnataka,West_Bengal,Uttar Pradesh]
#By Source Data

Bihar = list(func_execute(df_source_data,"Bihar","Registered murder cases"))
Delhi = list(func_execute(df_source_data,"Delhi","Registered murder cases"))
Uttar_Pradesh = list(func_execute(df_source_data,"Uttar Pradesh","Registered murder cases"))
Karnataka = list(func_execute(df_source_data,"Karnataka","Registered murder cases"))
West_Bengal = list(func_execute(df_source_data,"West Bengal","Registered murder cases"))

import matplotlib.pyplot as plt
fig,(ax1,ax2,ax3,ax4,ax5) = plt.subplots(nrows=5,ncols=1,figsize=(15,15))
ax1.pie(Bihar,labels =["2016","2017","2018"],shadow = True,autopct="%1.1f%%")
ax2.pie(Delhi,labels =["2016","2017","2018"],shadow = True,autopct="%1.1f%%")
ax3.pie(Uttar_Pradesh,labels =["2016","2017","2018"],shadow = True,autopct="%1.1f%%")
ax4.pie(Karnataka,labels =["2016","2017","2018"],shadow = True,autopct="%1.1f%%")
ax5.pie(West_Bengal,labels =["2016","2017","2018"],shadow = True,autopct="%1.1f%%")

ax1.set_xlabel("Bihar")
ax2.set_xlabel("Delhi")
ax3.set_xlabel("Uttar_Pradesh")
ax4.set_xlabel("Karnataka")
ax5.set_xlabel("West_Bengal")

plt.show()
```
